```python
# filtering to have only the required columns that needs the focus in STS data

def match_columns(df_columns, simplified_cols):
    matched_columns = {}
    for simplified in simplified_cols:
        for actual in df_columns:
            if simplified.lower().replace(' ', '') in actual.lower().replace(' ', ''):
                matched_columns[simplified] = actual
                break
    return matched_columns

# Use the function to get a dictionary of matched columns
matched_columns = match_columns(column_names_list_1, cols_df1)

# Assuming 'df' is your original DataFrame, filter to keep only matched columns
# Convert matched_columns values to a list since they are the actual column names in 'df'
df_filtered_1 = df1[matched_columns.values()]
```

```python
list_columns = ['ABP Mean', 'ART Mean', 'Core Temp', 'CVP', 'ECG HR', 'PA Mean', 'SpO2']

# Function to apply feature engineering to a column with lists of numbers
def create_summary_features(df, col_name):
    df[f'{col_name} mean'] = df[col_name].apply(lambda x: np.mean(x) if x else np.nan)
    df[f'{col_name} std'] = df[col_name].apply(lambda x: np.std(x, ddof=1) if x else np.nan)  # ddof=1 for sample standard deviation
    df[f'{col_name} min'] = df[col_name].apply(lambda x: min(x) if x else np.nan)
    df[f'{col_name} max'] = df[col_name].apply(lambda x: max(x) if x else np.nan)

# Apply feature engineering to each list column
for col_name in list_columns:
    create_summary_features(merged_df, col_name)

    # Optionally, drop the original column with lists if no longer needed
    merged_df.drop(col_name, axis=1, inplace=True)
```
